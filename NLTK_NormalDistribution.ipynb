```python
with open('chat.txt', "r") as file1:
    FileContent = file1.read()  
    text = FileContent
    print(FileContent) 
```

#Split into Words

```python
words = text.split()
print(words[:100])
```

#Removing Punctuations

```python
import string
table = str.maketrans('', '', string.punctuation)
stripped = [w.translate(table) for w in words]
print(stripped[:100])
```

#Convert to Lowercase

```python
words = [word.lower() for word in words]
print(words[:100])
```

**Now with NLTK all these can also be done. Implemented Below**




#Importing NLTK

```python
import nltk
nltk.download('all')
```

#Splitting into Sentences

```python
from nltk import sent_tokenize
sentences = sent_tokenize(text)
print(sentences[7])
```

#NLTK Splitting comma & periods become separate tokens 
eg - that's will be splitted into 'that' &  ','

```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens[:100])
```

#Lowercase conversion

```python
tokens_lower = [word.lower() for word in tokens] 
print(tokens_lower[:100])
```

#Filtering out Punctuations

```python
words = [word for word in tokens_lower if word.isalpha()]
print(words[:100])
```

#Filter out Stop Words

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
words = [w for w in words if not w in stop_words]
print(words[:100])
```

#Stemming - To root version

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
stemmed = [porter.stem(word) for word in words]
print(stemmed[:100])
```

#Lemmatization - To meaningful version

```python
from nltk.stem.wordnet import WordNetLemmatizer 
lem = WordNetLemmatizer()    
lemmed = [lem.lemmatize(word) for word in words]
print(lemmed[:100])
```

```python
print(stemmed)
```

#We will work with stem tokens

```python
fdist1 = nltk.FreqDist(stemmed)

filtered_word_freq = dict((word, freq) for word, freq in fdist1.items() if not word.isdigit())

print(filtered_word_freq)
```

#Sorting is descening order to list

```python
sort_orders = sorted(filtered_word_freq.items(), key=lambda x: x[1], reverse=True) 
print(sort_orders)
```

#list to dataframe

```python
import pandas as pd 
  

  
# Create the pandas DataFrame 
df = pd.DataFrame(sort_orders, columns = ['word', 'frequency']) 
  
# print dataframe. 
print(df )
```

```python
max = df['frequency'].max() 
max
```

#Standardize Dataframe

```python
df['frequency'] = df['frequency']/max 
print(df)
```

#Converting to csv

```python
df.to_csv("chat.csv")
```

#Dictionary to numpy array for simple plot

```python
# Python program to convert
# dictionary to numpy array
  
# Import required package
import numpy as np
  

  
# to return a group of the key-value
# pairs in the dictionary
result = filtered_word_freq.values()
  
# Convert object to a list
data = list(result)
  
# Convert list to an array
numpyArray = np.array(data)
  
# print the numpy array
print(numpyArray)
```
